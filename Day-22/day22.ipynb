# DAY 22 — BERT & Hugging Face

## Overview

In this notebook, we fine-tune a pretrained BERT model using Hugging Face Transformers for sentiment classification on the IMDB dataset.

## Install & Import Libraries

```python
import torch
import numpy as np
from datasets import load_dataset
from transformers import (
    BertTokenizerFast,
    BertForSequenceClassification,
    Trainer,
    TrainingArguments
)
```

## Load Dataset

```python
dataset = load_dataset("imdb")
dataset
```

## Tokenization

```python
tokenizer = BertTokenizerFast.from_pretrained("bert-base-uncased")

def tokenize(batch):
    return tokenizer(
        batch["text"],
        padding="max_length",
        truncation=True,
        max_length=256
    )

dataset = dataset.map(tokenize, batched=True)
dataset = dataset.remove_columns(["text"])
dataset = dataset.rename_column("label", "labels")
dataset.set_format("torch")
```

## Load Pretrained BERT

```python
model = BertForSequenceClassification.from_pretrained(
    "bert-base-uncased",
    num_labels=2
)
```

## Training Setup

```python
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=1)
    return {"accuracy": (preds == labels).mean()}

training_args = TrainingArguments(
    output_dir="./bert_results",
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=2,
    weight_decay=0.01,
    load_best_model_at_end=True,
    report_to="none"
)
```

## Train BERT

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset["train"],
    eval_dataset=dataset["test"],
    compute_metrics=compute_metrics
)

trainer.train()
```

## Evaluate

```python
trainer.evaluate()
```

## Observations

- BERT dramatically improves performance with minimal architecture changes.
- Fine-tuning is far more effective than training Transformers from scratch.
- Hugging Face abstracts complex training logic cleanly.
- This completes advanced NLP foundations
